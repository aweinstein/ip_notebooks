Compute the ratio between the volume of a $D$-dimensional sphere and a $D$-dimensional cube circumscribed to the sphere.

The volume of the $D$-dimensional sphere as a function of its radio $r$ is $ V_s(r) = \frac{\pi^{D/2}r^D}{\Gamma\left(1 + \frac{D}{2} \right)}$,where $\Gamma$ is the Gamma function. The volume of the $D$-dimensional cube as a function of the radius $r$ of the sphere is (the cube itself has edges of length $2r$) $V_c(r) = (2r)^D$.

```python
import numpy as np
from scipy.special import gamma
```

```python
D = np.arange(1, 21)
r = 1.
Vs = (np.pi**(D/2.) * r**D ) / gamma(1 + D/2.)
Vc = (2*r)**D
ratio = Vs / Vc
```

```python
plot(D, ratio, 'o-', lw=1.5)
ylim(ymax=1.05)
xlabel('Dimension $D$')
ylabel(r'$V_s/V_c$')
figure()
plot(D, Vs, 'o-', lw=1.5)
xlabel('Dimension $D$')
ylabel(r'$V_s$')
```
